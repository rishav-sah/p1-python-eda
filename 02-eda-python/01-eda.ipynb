```python
# Step 3: Create a dataframe using list

list1 = ["Tom", "Dick", "Harry"]
list2 = [31, 32, 33]
cols = ["name", "age"]
names = pd.DataFrame(zip(list1, list2), columns=cols)
```

```python
# How to create dataframe using dictionary

dict1 = {
    "names": ["Tom", "Dick", "Harry"],
    "age": [30, 31, 32]
}

# keys automatically becomes column
# value becomes row

pd.DataFrame(dict1)
pd.DataFrame(dict1, columns=["A", "B"]) # can't modify the column names

pd.DataFrame(dict1, index=["A", "B", "C"])

```
