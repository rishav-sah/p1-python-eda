#### Method 1: Map Method

- Read any categorical column: case_status
- check how many unique labels are there
- create a dictionary with those unique labels as keys by providing a number as a value

```python
df_visa['case_status'].unique()
```

```python
dict1 = {
    'Certified': 0,
    'Denied': 1
}
df_visa['case_status'].map(dict1)
```

```python
# do we want a overwrite column
# or existed column

df_visa['case_status_num'] = df_visa['case_status'].map(dict1)
```

```python
df_visa
```

```python
df_visa.drop('case_status_num', 
             axis=1, 
             inplace=True
            )
```

```python
df_visa['case_status'] = df_visa['case_status'].map(dict1)

# in the map method there's no inplace=True 
```

```python
df_visa
```

```python
list_of_continents = df_visa['continent'].unique()
```

```python
dict1 = {}
a = 0
for i in list_of_continents:
    dict1[i] = a
    a = a + 1
```

```python
df_visa['continent'] = df_visa['continent'].map(dict1)
```

```python
df_visa
```

```python
# using dictionary comprehension
labels = df_visa['continent'].unique()
num = len(df_visa['continent'].unique())
dict1 = {
    labels[i]:i for i in range(num)
}
dict1
```

```python
df_visa['continent'] = df_visa['continent'].map(dict1)
```

```python
df_visa
```

#### Method 2: $np.where()$

```python
# Read the data again
# np.where is applicable for binary conditions
# applicable for two labels
# np.where(<condition>,<True_value>, <False_value>)
# For example case_status have two labels
# condition: <col_values> == 'Certified'
# True value: Replace all certified values with 0
# False value: Replace all denied values with 1
```

```python
con = df_visa['case_status']=="Certified"
df_visa['case_status'] = np.where(con, 0, 1)
df_visa
```

#### One Hot Encoder

- one hot encoder means at a time only one will be ON(1/True), others are OFF(0/False)
- Suppose case status has two unique labels
- Certified and Denied
- In one hot encoder it will create two new columns
- like below

| case_status |  certified  |  denied|
|-------------|-------------|--------|
| certified | 1 | 0 |
| denied    | 0 | 0 |

**Advantage**

- One hot encoder new columns are orthogonal to each other
- Orthogonality means 90 degrees phase shift
- 90 degrees phase shift means perpendicular to each other
- Perpendicular means independent to each other
- Independent means no relation
- It is very important property that is to avoid relation between variables

**Drawbacks**
- Assume that you we have 100 unique labels for a column
- so it will create 100 new columns
- To process 100 columns we require more memory, more time and more hardward resources.
- This is called curse of dimensionality
