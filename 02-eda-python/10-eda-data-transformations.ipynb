## Data Transformations

- converting skewed distribution to normal distribution is very important
- The mathematics is developed by assuming as the data follows the normal distribution
- log transformations
- square root transformation
- exponential transformation
- box-cox transformation
- yeo-johnson transformation

```python
# step-1: read the packages
import numpy as np
import matplotlib.pyplot as plt

# step-2: read the data
data = np.random.exponential(size=1000)
# we are taking a random values from exponential distribution
# we are taking 1000 samples
len(data)


plt.hist(data, bins=40)
plt.title("Exponential Data")
plt.figure(figsize=[10, 5])
plt.show()
```

#### Log Transformation

- np.log represents natural logarithm  (base e)
- exponential will multiply with log base e
- natural logarithms will work on positive data
- log transformation will remove the skew
- It will not covert into normal distribution

```python
plt.hist(np.log(data), bins=40)
plt.title("Log Transformation")
plt.figure(figsize=[10, 5])
plt.show()
```

```python
plt.subplot(1,2,1).hist(data, bins=40)
plt.subplot(1,2,2).hist(np.log(data), bins=40)
plt.show()
```
