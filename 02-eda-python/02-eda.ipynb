```python
# task 1: seperate categorial and numerical columns
```

```python
df_visa_dict = df_visa.dtypes.to_dict()
df_visa_dict
```

```python
# create two lists
df_visa_cat = []
df_visa_num = []
for key, value in df_visa_dict.items():
    print(f"{key}: {value}")
    if (value == "object"):
        df_visa_cat.append(key)
    else:
        df_visa_num.append(key)


print("The categorial columns (df_visa_cat):", df_visa_cat)
print("The numerical columns (df_visa_num):", df_visa_num)
```

```python
# better way - list comprehension

df_visa_cat = [key for key, value in df_visa_dict.items() if value == "object"]
df_visa_num = [key for key, value in df_visa_dict.items() if value == "int64"]

print("The categorical columns: ", df_visa_cat)
print("The categorical columns: ", df_visa_num)
```

#### $take$

```python
# retrieving specific rows
list1 = [100, 200, 300] # 100th row, 200th row, 300th row
df_visa.take(list1)
```

```python
list1 = [1, 2, 11]
df_visa.take(list1, axis=1)

# in python the index starts with 0
# 1 ====== 2nd column
# 2 ====== 3rd column
# 11 ===== 12th column

# default axis = 0
```

```python
list2 = [1, 2, 12] # 12th column not available in the dataframe. error code IndexError: indices are out-of-bounds
df_visa.take(list2, axis=1)
```

##### $Task-2$

```python
# Retrieving 150, 300, 450 rows from 5, 8 and 12

list1 = [150, 300, 450]
list2 = [4, 7, 11]

df_visa_temp = df_visa.take(list1, axis=0)
df_visa_final = df_visa_temp.take(list2, axis=1)
df_visa_final
```

```python
df_visa.take([150, 300, 450], axis=0).take([5, 8, 11], axis=1) ## as both the methods returns the dataframe
df_visa.take([4, 7, 11], axis=1).take([150, 300, 450]) # works vice versa but better to decide the column first then the row
```
