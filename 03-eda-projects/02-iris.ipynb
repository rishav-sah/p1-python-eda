```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset
file_path = "./dataset/iris/Iris.csv"
df = pd.read_csv(file_path)
df
```

```python
# Initial Inspection to start getting the dataset
print("Shape: ")
print(df.shape)

print("------ Columns --------")
print(df.columns)

print("------ Missing value counts --------")
print(df.isnull().sum())

print("------ Sample rows --------")
print(df.head())

print("------ Datatypes --------")
print(df.dtypes)
```

```python
# Overall preview
# In the overall preview, we can observe that it consists overall information of the dataframe
print(df.info())
```

```python
# checking individual datatypes count
df.dtypes.value_counts()
```

```python
# includes all numeric data types (int64, float64)
# If the dataset is huge then we can categorize more on depending on the data types
df_num = df.select_dtypes(include=[np.number])
df_cat = df.select_dtypes(exclude=[np.number])

print("Numerical Columns")
print(df_num.columns.to_list())

print("Categorical Columns")
print(df_cat.columns.to_list())
```

```python
#### Categorical Column Analysis

species_count = df_cat['Species'].value_counts()
print(species_count)
df_species = pd.DataFrame(zip(species_count.keys(), species_count.values), columns=['species', 'count'])
df_species
```

```python
plt.figure(figsize=[8, 5])
plt.title("Bar plot of different species in the Iris Flower")
plt.xlabel("Species")
plt.ylabel("Count")
plt.bar("species", "count", data=df_species)
plt.show()
```

#### Iris Dataset Insights

##### Basic information about the Iris Flower

- The Iris flower is a popular and well-known plant, known for its beauty, symmetry, and shape.
- The flower has three upright “standards” (petals), and three drooping “falls” (modified leaves or sepals).
- Iris comes in a variety of colors and species.


##### Some insights as per the dataset
- The dataset contains 150 rows and 6 columns.
- The numerical columns represent the length and width of the sepal and petal respectievely, measured in centimeters. Each observation is identified by a unique ID.
- The dataset consists of 1 categorical column named `Species`, that contains three unique values: setosa, versicolor, and virginica.
- The species consists of three unique sub-category or nominal columns.
- The values in the `species` column are nominal categories, meaning they represent distinct groups without any inherent order.
- In the dataset, each value of `species` are equally represented as seen in the bar plot, with 50 samples per species.
